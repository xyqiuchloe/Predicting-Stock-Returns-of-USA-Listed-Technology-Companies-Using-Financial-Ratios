# Impute missing value

```python
dff = pd.read_csv("/Users/xinrucheng/Desktop/capstone/dfv1.csv") 
dff.head()
```

```python
dff['Cost.of.Revenue'].describe()
```

# Find correlation

```python
#find the columns that has correlation with Operating Income larger than 0.7
corr = dff.corr()
abs(corr[['Cost.of.Revenue']])[abs(corr['Cost.of.Revenue']) > 0.8].sort_values(by = 'Cost.of.Revenue', ascending = False )
#get those columns' names
names = abs(corr[['Cost.of.Revenue']])[abs(corr['Cost.of.Revenue']) > 0.8].sort_values(by = 'Cost.of.Revenue', ascending = False ).index.values.tolist()
names
#Pretax Income(Loss) has the highest correlation
#check if it has missing value in rows where Operating Income is null: 100% non-missing
dff[dff['Cost.of.Revenue'].isnull()]['Revenue'].isnull().any()
```

```python
#check the linearity: pretty linear  
mydata = dff[["Cost.of.Revenue", "Revenue"]].dropna(how="any")
mydata.plot(kind= 'scatter', x="Revenue", y="Cost.of.Revenue")
plt.xlabel("Revenue")
plt.ylabel("Cost.of.Revenue")
plt.show()
```

```python
#fit linear regression and use the fitted value to fill missing value
regressor = LinearRegression()  
train = dff[["Cost.of.Revenue","Revenue" ]][dff["Cost.of.Revenue"].notnull()]
regressor.fit(train.iloc[:,0].values.reshape(-1,1), train.iloc[:,1].values.reshape(-1,1)) #training the algorithm
predict1 =  regressor.predict(dff["Revenue"].values.reshape(-1,1))
```

```python
#check result 
plt.plot(dff["Revenue"].values.reshape(-1,1), predict1, color='red', linewidth=2)
plt.scatter(mydata["Revenue"], mydata["Cost.of.Revenue"])
plt.show()
```

```python
dff.loc[dff["Cost.of.Revenue"].isnull(),"Cost.of.Revenue"]=list(predict1.reshape(-1)[np.array(dff["Cost.of.Revenue"].isnull())])
```

```python
dff['Cost.of.Revenue'].isna().sum() #done!!
```
